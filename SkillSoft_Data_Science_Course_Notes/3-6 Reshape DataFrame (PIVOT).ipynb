```python
import pandas as pd 
```

```python
avocados = pd.read_csv('DataSets/avocado.csv')
avocados.head()
```

```python

# How to know the unique values into the data set

avocados['region'].unique()

```

```python


# Limitin the large of dataset to only few entries. 

avocados_totalUS = avocados[(avocados['region'] == 'TotalUS') & 
                            (avocados['Date'] > '2018-01-01')]
avocados_totalUS
```

```python
# Sort the data

avocados_totalUS.sort_values(['Date'])
```

```python

# CReating a Pivot Table for better analysis and better view of the data. 


avocados_pivot = avocados_totalUS.pivot(index = 'Date', # this will be the rows unique
                                        columns = 'type',  # will show the values of the variable as columns
                                        values = 'AveragePrice') # will fill the table with those values. 

avocados_pivot


'''The convination of Index and Colums must be unique to generate a pivot table. '''
```
