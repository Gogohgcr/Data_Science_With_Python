```python
import pandas as pd
import numpy as np 
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
politicians_df = pd.read_csv('DataSets/Politicians.csv')
politicians_df
```

```python
# Add a new column in the data frame

# You can specify an specific number or a list 
politicians_df['Year of Bird'] = 1900
politicians_df

politicians_df['Year of Bird'] = [1881,1882,1874,1790,1789,1918,1904]
politicians_df


# Or set an array to do so 
dates = np.array ([1,2,3,4,5,6,7])

politicians_df['Year of Bird'] = dates
politicians_df
```

```python
# Removing a column from the dataframe. 


del politicians_df['Year of Bird']

politicians_df

# we can confir that the columnd was deleted properly. 
```

```python
# Adding a new row in the DataFrame. 

politicians_df.loc[-1] = ['Otto von Bismarck', 'Germany']
#-1 is the last space of the rows.
politicians_df

```

```python
# Modify the index of the dataFrame

politicians_df.index = politicians_df.index +1
politicians_df

# This sum a number each of the index but it is not sorted so we can do so

```

```python

#SORT

politicians_df = politicians_df.sort_index()
politicians_df
```

```python
# Drop a Row in the DF

politicians_df = politicians_df.drop(politicians_df.index[7])
politicians_df
```

```python
# Set new index values 

index_values = [1815,1881,1882,1874,1890,1889,1918]

politicians_df.index = index_values 

politicians_df
```

```python
# Also you can revert the changes
# Pandas offer the option with reset_index. 


politicians_df = politicians_df.reset_index()
politicians_df

```

```python
# this will create a column named index so it need to be deleted


politicians_df = politicians_df.drop(columns= ['index'])
politicians_df
```
