```python
import pandas as pd 
```

```python
# Here we are importing the data set and playing with the multi-indexing
# We are selecting as index the region and the date. 


avocados = pd.read_csv('DataSets/avocado.csv',
                      index_col = ['region','Date'])

avocados= avocados.sort_index()

avocados
```

```python
# Filtering the data to only orgnic values 

avocados= avocados[avocados['type']== 'organic']
avocados

```

```python


#Apply the filter in the Index based on the value filds
# if can be done using the index.get_level_values 

avocados= avocados.loc[(avocados.index.get_level_values ('Date') > '2018-03-01')]
avocados




```

```python
'''THe index of this data frame is a multIndex due to the amount of indexes'''

avocados.index
```

```python
avoca = pd.read_csv('DataSets/avocado.csv')

avocados_totalUS = avoca[(avoca['region'] == 'TotalUS') & 
                            (avoca['Date'] > '2018-01-01')]
avocados_totalUS


avocados_pivot = avocados_totalUS.pivot(index = 'Date', # this will be the rows unique
                                        columns = 'type',  # will show the values of the variable as columns
                                        values = 'AveragePrice') # will fill the table with those values. 

avocados_pivot
```

```python
#Stack operation 

#Stacks the content of a single row into multiple rows. 
# in this cass

avocados_stack = avocados_pivot.stack()
avocados_stack
```
