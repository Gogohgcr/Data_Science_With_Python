```python
# Subseting the DF with ILOC
#Show the rows from 2 to 6th index with all the columns

towed_cars.iloc[2:6]
```

# Subseting with conditions

```python
# Show all those values from the cars that their year is greatest than 2005

towed_cars[towed_cars['Vehicle_Year']> 2005]
```

```python
#Returs the 

towed_cars[towed_cars['Vehicle_State'] ==  'CT']
```

```python
# Applying both conditions with AND operator &

towed_cars[(towed_cars['Vehicle_Year']> 2005) & 
           (towed_cars['Vehicle_State'] == 'CT')]
```

```python
# Applying both conditions with OR operator |

towed_cars[(towed_cars['Vehicle_Year']> 2005) |
           (towed_cars['Vehicle_State'] == 'CT')]
```

```python

# NEGATIONS CONDITIONS ~

#Show all those vehicles that the color is NOT  BLACK

towed_cars[~(towed_cars['Color'] == "BLACK")]
```
